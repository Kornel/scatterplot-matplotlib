```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
test_df = pd.DataFrame({
        "X": [1, 2, 3, 4],
        "Y": [5, 4, 2, 1],
        "C": [1, 2, 3, 4]
    })
```

# All nice and clean

```python
test_df.plot(kind="scatter", x="X", y="Y", s=50);
```

# Where are my X axis labels?! ðŸ˜¡

```python
test_df.plot(kind="scatter", x="X", y="Y", c="C");
plt.tight_layout();
```

```python
test_df.plot(kind="scatter", x="X", y="Y", s=50, c="C", cmap="plasma");
plt.tight_layout();
```

# Use the force?

```python
ax = test_df.plot(kind="scatter", x="X", y="Y", s=50, c="C", cmap="plasma");
ax.set_xticklabels([0, 1]);
ax.set_xticks([0, 1]);
plt.tight_layout();
```

# set_xlabel does not change anything

```python
ax = test_df.plot(kind="scatter", x="X", y="Y", c="C");
ax.set_xlabel("X");
plt.show();
```

# But subplots does!

```python
fig, ax = plt.subplots()
test_df.plot(kind="scatter", x="X", y="Y", s=50, c="C", cmap="plasma", ax=ax);
```
